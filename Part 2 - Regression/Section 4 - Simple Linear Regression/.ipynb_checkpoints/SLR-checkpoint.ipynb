```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

%matplotlib notebook
```

### Load Data

```python
salary = pd.read_csv("Salary_Data.csv")
salary.head()
```

### Feature Selection

Reshape both `X` and `y` into column vectors.

```python
X = salary.YearsExperience.values.reshape(-1, 1)
y = salary.Salary.values.reshape(-1, 1)
```

### Training and Test Set Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3,
                                                    random_state=123)
```

### Model Training

```python
slr = LinearRegression()
slr.fit(X_train, y_train)
```

### Model Prediction

```python
train_pred = slr.predict(X_train)
```

```python
plt.scatter(X_train, y_train, color="red")
plt.plot(X_train, train_pred, color="blue")

plt.title("Salary against Years of Experience (on training set)")
plt.xlabel("Years of experience")
plt.ylabel("Salary ($)")
plt.show()
```

### Model Evaluation

```python
test_pred = slr.predict(X_test)
```

```python
plt.scatter(X_train, y_train, color="red")
plt.scatter(X_test, y_test, color="green")
plt.plot(X_train, train_pred, color="blue")

plt.title("Salary against Years of Experience (on both sets)")
plt.xlabel("Years of experience")
plt.ylabel("Salary ($)")
plt.show()
```
