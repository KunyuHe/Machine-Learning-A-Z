```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from collections import Counter
```

### Load Data

```python
mall = pd.read_csv("Mall_Customers.csv")
mall.head()
```

### Data Cleaning

```python
mall.isnull().sum(axis=0)
```

No value missing.

### Feature Selection

```python
X = mall.iloc[:, 3:5].values.astype(float)
X.shape
```

### Model Training

```python
def find_best_k(ax, X, max_k=10, seed=123, insert_vertical_at=None):
    wcss = []
    
    for i in range(1, max_k+1):
        kms = KMeans(n_clusters=i, init='k-means++', random_state=123)
        kms.fit(X)
        wcss.append(kms.inertia_)
    
    ax.plot(range(1, max_k+1), wcss)
    ax.set_title("KMeans Elbow: k from 1 to {}".format(max_k),
                 fontdict={'fontsize': 15, 'fontweight': "bold"})
    ax.set_xlabel("Number of clusters", fontsize=13)
    ax.set_ylabel("Within-cluster Sum of Squares (WCSS)", fontsize=13)
    
    ax.set_xticks(range(1, max_k+1))
    ax.get_xaxis().set_tick_params(direction='in', length=3)
    
    ax.set_yticks([])
    
    if insert_vertical_at:
        ax.axvline(x=insert_vertical_at, color='r', linestyle='--')
        print("The best number of clusters is {}.".format(insert_vertical_at))
```

```python
fig = plt.figure(figsize=[12, 8])
ax = fig.add_subplot(1, 1, 1)

find_best_k(ax, X, max_k=20, insert_vertical_at=5)
plt.show()
```

```python
kms = KMeans(n_clusters=5, init='k-means++', random_state=123)
y_kmeans = kms.fit_predict(X)
Counter(y_kmeans)
```

```python
fig = plt.figure(figsize=[12, 8])

plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], c="red", alpha=0.8,
            edgecolors="black", label="Standard")
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], c="blue", alpha=0.8,
            edgecolors="black", label="Target")
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], c="green", alpha=0.8,
            edgecolors="black", label="Careful")
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], c="cyan", alpha=0.8,
            edgecolors="black", label="Sensible")
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], c="magenta", alpha=0.8,
            edgecolors="black", label="Careless")

plt.title("Customers Segmented by KMeans",
          fontdict={'fontsize': 15, 'fontweight': "bold"})
plt.xlabel("Annual Income (k$)", fontsize=13)
plt.ylabel("Spending Score (1-100)", fontsize=13)
plt.legend(fontsize=12)

plt.show()
```
